```python
# 18. Write a Python program to generate all permutations of a list in Python.
# Way to achive the list permutations
# 1. itertools build in methods
import itertools

list18=[90,1,2,3]

lr18=list(itertools.permutations(list18))

print('Possible permutations of list:{} are as below'.format(list18))

for i in lr18:
  print(i)

#without using build in method

def list_permut(lst):
  if len(lst)==0:
    return[[]]
  perm=[]
  for i in range(len(lst)):
    elem=lst[i]
    rem_ele=lst[:i]+lst[i+1:]
    for p in list_permut(rem_ele):
      perm.append([elem]+p)
  return perm

print('Possible permutations using looping of list:{} are as below'.format(list18))
for i in list_permut(list18):
  print(i)



```

```python
# 26. Write a python program to check whether two lists are circularly identical.

def check_circ_identical(lst1,lst2):
  if(len(lst1)!=len(lst2)):
    return False

  for x in range(len(lst1)):
    new_list=lst1[x:]+lst1[:x]
    if new_list==lst2:
      return True
  return False

list26_1=[10,20,30,40]
list26_2=[30,40,20,10]

flag=check_circ_identical(list26_1,list26_2)
if flag:
  print('Both list are circularly identical')
else:
  print('Both list are not circularly identical')
```

```python
# 34. Write a Python program using Sieve of Eratosthenes method for computing primes upto a specified number.
#  Ἐ Note: In mathematics, the sieve of Eratosthenes, (Ancient Greek: κόσκινον ρατοσθένους, kóskinon Eratosthénous)
#  one of a number of prime number sieves, is a simple, ancient algorithm for finding all prime numbers up to any given limit.

def isPrime(num):
    if num < 2:
        return False

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False

    return True


def sieve(n):
    res = []

    for i in range(2, n + 1):
        if isPrime(i):
            res.append(i)

    return res


n=int(input('Enter the number:'))
res = sieve(n)

for ele in res:
  print(ele, end=' ')
```
