```python
#Write a Python program to check whether a year is leap year or not.
#Take year as input from user

year=int(input('Enter year:'))

d4=year%4
d100=year%100
d400=year%400

print('Using if else statement-')
if(d4==0 and d100==0 and d400==0):
  print('Year: ',year,' is leap year')

else:
  print('Year: ',year,' is not leap year')

#using clander method

import calendar

yearc=calendar.isleap(year)
print('Using calender methods -')
if(yearc):
  print('Year: ',year,' is leap year')

else:
  print('Year: ',year,' is not leap year')
```

```python
#Write a Python program to check whether a character is alphabet or not.

char=str(input('Enter any string:'))


for s in char:
  t=s.isalpha()
  if(t):
    print(s,' is alphabet')
  else:
    print(s," is not alphabet")
```

```python
#Write a Python program to input any alphabet and check whether it is
#vowel or consonant

char1=str(input('Enter any char:'))

vowels=['a','A','e','E','i',"I",'o','O','u','U']

print('Using simple if-else statement')

if(char1=='a' or char1=='A' or char1=='E' or char1=='e' or char1=='i' or char1=='I' or char1=='o' or char1=='O' or char1=='u' or char1=='U'):
  print('Char:',char1,' is vowel')

else:
  print('Char: ',char1, ' is consonant')

print('Using another logic ')

i=vowels.count(char1)

if(i>0):
   print('Char:',char1,' is vowel')

else:
  print('Char: ',char1, ' is consonant')


```
